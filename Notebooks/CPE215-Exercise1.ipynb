### Import all the needed libraries

```python
import random
```

### Define our toy linear problem
##### A =14 , C = 19 , B = 5 -> 14X + 5Y + 19

```python
def function(x ,y):
    return 14*x + 5*y + 19
```

## Define Useful Variables

```python
NUMBER_OF_SAMPLES = 150
```

## Generate data and labels
Note: 150 random samples (0-1)

```python
x = [random.random() for _ in range(NUMBER_OF_SAMPLES)]
y = [random.random() for _ in range(NUMBER_OF_SAMPLES)]
z = [function(x_, y_) for x_, y_ in zip(x,y)]
```

## Combining the data through zip

```python
data = list(zip(x,y,z))
data[0:10]
```

## Initialize the parameters of the linear regressor

```python
m1 = random.random()
m2 = random.random()
b = random.random()
```

## Define a linear rate for a more stable gradient

```python
lr = 0.0055
```

## Define a function for getting the resut of the regressor

```python
def get_answer(x,y):
    return m1*x+m2*y+b
```

## Verify the linear regressor
### Untrained

```python
x_,y_,z_ = data[0]
print(get_answer(x_,y_), z_)
print ("AME:",abs(get_answer(x_,y_)- z_))
```

## Define the loss function for the regressor

```python
def get_error(x,y,z):
    P = get_answer(x,y)
    error = abs(P-z)
    return error
```

## Verify error function:


```python
get_error(x_,y_,z_)
```

## Function definition for calculating gradients:

```python
def update_parameters(x,y,z):
    global m1,m2,b
    #Gradinet computation  
    #gradients are calculated at 1 data per update
    factor=((get_answer(x,y) - z))/abs(get_answer(x,y) - z)
    dm1 = factor*x
    dm2 = factor*y
    db = factor
    #New values
    m1 = m1 - dm1 * lr
    m2 = m2 - dm2 * lr
    b = b - db* lr
```

## Update gradient throughout all of the data and show the learned parameters and error

```python
count =0

#Iterate through the data 100 times.
for _ in range(100):
       for x_, y_, z_ in data:
        count = count+1
        error = get_error(x_,y_, z_)
        print(count,':','x: ',x_, ' y: ',y_,' z: ',z_)
        print('\tPrediction:', get_answer(x_,y_))
        print('\tError:', error)
        update_parameters(x_,y_,z_)
       
```
